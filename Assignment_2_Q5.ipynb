```python
import numpy as np

X = np.array([[1,2],[1,4],[1,5]]).reshape(3,2)
beta = np.array([0,1]).reshape(2,1)
y = np.array([1.2,2.8,5.3]).reshape(3,1)



def gradient_func(beta,grad_beta_loss,n):
    for i in range(n+1):
        updated_beta = np.round(beta - 0.02*grad_beta_loss,4)
        print(f"beta{i} =\n {updated_beta}")
        print("")
        beta = updated_beta
        
        
n = int(input("Enter number of iterations = "))

grad_beta_loss = 2*(X.T@X@beta-X.T@y)

gradient_func(beta,grad_beta_loss,n)
```
