```python
import numpy as np
```

```python
def sigmoid(x):
  return 1/(1+np.exp(-x))

def dsigmoid(x):
  return sigmoid(x)*(1-sigmoid(x))
```

```python
def a_1(w1, x):
  return sigmoid(w1*x)
```

```python
def yhat(w2, a1):
  return sigmoid(w2*a1)
```

```python
def loss(y, yhat):
  return (y-yhat)**2
```

```python
def gradw1(y, w2, w1, x):
  a1 = a_1(w1, x)
  yh = yhat(w2, a1)
  return -2*(y-yh)*dsigmoid(w2*a1)*w2
```

```python
def gradw2(y, w2, w1, x):
  a1 = a_1(w1, x)
  yh = yhat(w2, a1)
  return -2*(y-yh)*dsigmoid(w2*a1)*a1
```

```python
x = 1
w1 = 1
w2 = 2
y = 0.5
```

```python
yh = yhat(w2, a_1(w1, x))

print("Loss: ", loss(y, yh))
```

Gradient Descent

```python
eta = 0.5

for i in range(10000):
  w1 = w1 - eta*gradw1(y, w2, w1, x)
  w2 = w2 - eta*gradw2(y, w2, w1, x)
  yh = yhat(w2, a_1(w1, x))
  if i%100:
    print("yhat = ", yh, "Loss: ", loss(y, yh))
```
