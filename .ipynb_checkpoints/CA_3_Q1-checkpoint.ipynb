```python
import numpy as np
import matplotlib.pyplot as plt

# Inside the stats module from scipy import norm for normalization
from scipy.stats import norm

def simulate_die_throws(n):
    # Simulate n throws of a die
    throws = np.random.randint(1, 7, size=(n,))

    # Calculate sum, mean, and variance
    total_sum = np.sum(throws)
    mean_value = np.mean(throws)
    variance_value = np.var(throws)

    # Plot histogram
    plt.hist(throws, bins=range(1, 8), align='left', density=True, rwidth=0.8, edgecolor='black')
    plt.title(f'Histogram of {n} Die Throws')
    plt.xlabel('Die Face')
    plt.ylabel('Probability')
    plt.show()

    # Plot Gaussian distribution based on mean and variance
    x = np.linspace(0.5, 6.5, 100)
    plt.plot(x, norm.pdf(x, mean_value, np.sqrt(variance_value)),'.g', label='Gaussian Distribution')
    plt.title(f'Gaussian Distribution for {n} Die Throws')
    plt.xlabel('Sum of Die Throws')
    plt.ylabel('Probability Density')
    plt.legend()
    plt.show()

    return total_sum, mean_value, variance_value

# Example usage
n_values = [10, 50, 100]
for n in n_values:
    total_sum, mean_value, variance_value = simulate_die_throws(n)
    print(f"For {n} throws: Sum = {total_sum}, Mean = {mean_value:.2f}, Variance = {variance_value:.2f}")
```
